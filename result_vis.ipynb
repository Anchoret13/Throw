```python
import matplotlib.pyplot as plt
import numpy as np

file_name = './logging/recordings/name_Throwing__noise_0.02__agent_usher.txt'
f = open(file_name, encoding = "utf-8")
result = f.read()
result = result.split("\n")
result = result[:-1]
for i in range(len(result)):
    result[i] = result[i].split(",")
    for j in range(len(result[0])):
        result[i][j] = float(result[i][j])
```

```python
result
```

```python
num_colum = []
success_rate = []
reward = []
value = []

for i in range(len(result)):
    num_colum.append(int(result[i][0]))
    success_rate.append(result[i][1])
    reward.append(result[i][2])
    value.append(result[i][3])
    
fig = plt.figure(num = 1, figsize = (10, 10))
plt.plot(num_colum, reward)
```
